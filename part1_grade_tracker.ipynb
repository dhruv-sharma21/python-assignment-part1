# Student Grade Tracker

## Task 1 — Data Parsing & Profile Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for s in raw_students:

    # 1. stripping whitespace first then title case for proper formatting
    clean_name = s["name"].strip()
    clean_name = clean_name.title()

    # roll was a string so converting to int
    roll_number = int(s["roll"])

    # splitting marks string on ", " to get individual marks
    marks_string = s["marks_str"]
    marks_list_str = marks_string.split(", ")

    marks = []
    for m in marks_list_str:
        marks.append(int(m))   # converting each mark from string to int

    clean_student = {
        "name": clean_name,
        "roll": roll_number,
        "marks": marks
    }

    cleaned_students.append(clean_student)

    # 2. checking every word - if any word has non-alphabet char, name is invalid
    words = clean_name.split()

    is_valid = True
    for word in words:
        if word.isalpha() == False:
            is_valid = False

    if is_valid == True:
        name_status = "✓ Valid name"
    else:
        name_status = "✗ Invalid name"

    # 3. printing the profile card for each student
    print("================================")
    print(f"Student : {clean_student['name']}   {name_status}")
    print(f"Roll No : {clean_student['roll']}")
    print(f"Marks   : {clean_student['marks']}")
    print("================================")
    print("")

# 4. looping through to find roll 103 and printing name in both cases
for student in cleaned_students:
    if student["roll"] == 103:
        print(f"ALL CAPS  : {student['name'].upper()}")
        print(f"lowercase : {student['name'].lower()}")
```

## Task 2 — Marks Analysis Using Loops & Conditionals

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# 1. going through each subject and assigning grade based on marks range
for i in range(len(subjects)):
    m = marks[i]

    if m >= 90 and m <= 100:
        grade = "A+"
    elif m >= 80 and m <= 89:
        grade = "A"
    elif m >= 70 and m <= 79:
        grade = "B"
    elif m >= 60 and m <= 69:
        grade = "C"
    else:
        grade = "F"

    print(f"{subjects[i]} : {m} : {grade}")

print("")

# 2. manually adding up marks instead of using sum()
total = 0
for m in marks:
    total = total + m

average = round(total / len(marks), 2)

# starting with first element and replacing if something higher is found
highest_marks = marks[0]
highest_subject = subjects[0]
for i in range(len(marks)):
    if marks[i] > highest_marks:
        highest_marks = marks[i]
        highest_subject = subjects[i]

# same logic for lowest
lowest_marks = marks[0]
lowest_subject = subjects[0]
for i in range(len(marks)):
    if marks[i] < lowest_marks:
        lowest_marks = marks[i]
        lowest_subject = subjects[i]

print(f"Total   : {total}")
print(f"Average : {average}")
print(f"Highest : {highest_subject} ({highest_marks})")
print(f"Lowest  : {lowest_subject} ({lowest_marks})")
```

```python
# 3. while loop keeps running until user types done
new_subjects_added = 0

while True:
    subject_input = input("Enter subject name (or done to stop): ")

    if subject_input == "done":
        break

    marks_input = input(f"Enter marks for {subject_input} (0-100): ")

    # checking if input is actually a number before converting
    if marks_input.isnumeric() == False:
        print("warning: that is not a valid number, skipping")
        continue

    marks_value = int(marks_input)

    # making sure marks are within the valid range
    if marks_value < 0 or marks_value > 100:
        print("warning: marks should be between 0 and 100, skipping")
        continue

    subjects.append(subject_input)
    marks.append(marks_value)
    new_subjects_added = new_subjects_added + 1   # only counts valid entries

print(f"\nNew subjects added : {new_subjects_added}")

updated_total = 0
for m in marks:
    updated_total = updated_total + m

updated_average = round(updated_total / len(marks), 2)
print(f"Updated average    : {updated_average}")
```

## Task 3 — Class Performance Summary

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# storing averages separately so i can reuse them later
averages = []

# 1. calculating average for each student
for student in class_data:
    name = student[0]
    marks = student[1]

    total = 0
    for m in marks:
        total = total + m

    avg = round(total / len(marks), 2)
    averages.append(avg)

# 2. printing the table - :<18 keeps columns aligned
print(f"{'Name':<18}| Average | Status")
print("-" * 40)

passed = 0
failed = 0

for i in range(len(class_data)):
    name = class_data[i][0]
    avg  = averages[i]

    if avg >= 60:
        status = "Pass"
        passed = passed + 1
    else:
        status = "Fail"
        failed = failed + 1

    print(f"{name:<18}|  {avg:<7}| {status}")

# 3. summary after the table
print("")
print(f"Passed : {passed}")
print(f"Failed : {failed}")

# same approach as task 2 - start with first student and update if better found
topper_name = class_data[0][0]
topper_avg  = averages[0]

for i in range(len(averages)):
    if averages[i] > topper_avg:
        topper_avg  = averages[i]
        topper_name = class_data[i][0]

print(f"Class Topper  : {topper_name} ({topper_avg})")

total_avg = 0
for a in averages:
    total_avg = total_avg + a

class_average = round(total_avg / len(averages), 2)
print(f"Class Average : {class_average}")
```

## Task 4 — String Manipulation Utility

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. strip first - all other steps use clean_essay
clean_essay = essay.strip()
print("1. Stripped Essay:")
print(clean_essay)
print("")

# 2. title() capitalises first letter of every word
title_essay = clean_essay.title()
print("2. Title Case:")
print(title_essay)
print("")

# 3. clean_essay is already lowercase so no need for .lower() before count
python_count = clean_essay.count("python")
print("3. Count of python:")
print(python_count)
print("")

# 4. replace all occurrences of python with the emoji version
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("4. After Replacing python:")
print(replaced_essay)
print("")

# 5. splitting on ". " breaks the essay into separate sentences
sentences = clean_essay.split(". ")
print("5. Sentences List:")
print(sentences)
print("")

# 6. adding a dot at the end if sentence doesnt already have one
print("6. Numbered Sentences:")
for i in range(len(sentences)):
    sentence = sentences[i]

    if sentence.endswith(".") == False:
        sentence = sentence + "."

    print(f"{i + 1}. {sentence}")
```
